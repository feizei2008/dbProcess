```python
import pandas as pd
from pymongo import MongoClient
from bson.son import SON
def get_Items(db, coll_name):
    Items = db[coll_name].find({}) # 注：原始db的datetime为ISO格式需要转换才能比较
    return Items

client = MongoClient("localhost",27017)
db = client["VnTrader_Tick_Db"]
# items = db["rb1901"].find({}).limit(3)
# for i in items:
#     print i 
pipeline = [{"$group": {"_id": "$date", "count": {"$sum": 1}}},
            {"$sort": SON([("date", -1), ("_id", -1)])}] # 按日期倒序排序，也可按count排，等等
# nums = list(db[coll_name].aggregate(pipeline))
item_all = get_Items(db,'rb1901')
coll = db['rb1901']
dates = list(coll.aggregate(pipeline))
dateRange = [i["_id"] for i in dates]
print dateRange

def get_Oneday_Items(db, coll_name, date):
    Items = db[coll_name].find({"date": {'$eq': date}}) # 注：原始db的datetime为ISO格式需要转换才能比较
    return Items

db20180906 = get_Oneday_Items(db, 'rb1901', '20180906')
df10 = pd.DataFrame(list(db20180906))
# print len(df10)
# df11 = df10
# df12 = df10.append(df11)
# print len(df12)

for i in dateRange:
    items_daily = get_Oneday_Items(db, 'rb1901', i)       
    df = pd.DataFrame(list(items_daily))
    df666 = df10.append(df)
print len(df666)        
```
